## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# Scatter plot of Age and WorkWeekHrs
scatter_query = """
SELECT Age, WorkWeekHrs
FROM master
"""

scatter_data = pd.read_sql_query(scatter_query, conn)

plt.scatter(scatter_data['Age'], scatter_data['WorkWeekHrs'])
plt.title('Nuage de points entre âge et heures de travail par semaine')
plt.xlabel('Âge')
plt.ylabel('Heures de travail par semaine')
plt.show()

# Calcul du coefficient de corrélation entre Age et WorkWeekHrs
correlation_coefficient = scatter_data['Age'].corr(scatter_data['WorkWeekHrs'])

print(f"Le coefficient de corrélation entre Age et WorkWeekHrs est : {correlation_coefficient}")

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# Bubble plot of WorkWeekHrs and CodeRevHrs
bubble_query = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""

bubble_data = pd.read_sql_query(bubble_query, conn)

plt.scatter(bubble_data['WorkWeekHrs'], bubble_data['CodeRevHrs'], s=bubble_data['Age']*2, alpha=0.5)
plt.title('Diagramme à bulles entre heures de travail et heures de révision de code')
plt.xlabel('Heures de travail par semaine')
plt.ylabel('Heures de révision de code')
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# Stacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35
stacked_chart_query = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

stacked_chart_data = pd.read_sql_query(stacked_chart_query, conn)

stacked_chart_data.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median().plot(kind='bar', stacked=True)
plt.title('Diagramme empilé des médianes d\'heures de travail et d\'heures de révision par groupe d\'âge 30-35')
plt.xlabel('Âge')
plt.ylabel('Médiane')
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# Line chart of median ConvertedComp for all ages from 45 to 60
line_chart_query = """
SELECT Age, AVG(ConvertedComp) as MedianConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
GROUP BY Age
"""

line_chart_data = pd.read_sql_query(line_chart_query, conn)

plt.plot(line_chart_data['Age'], line_chart_data['MedianConvertedComp'], marker='o')
plt.title('Graphique des médianes de ConvertedComp pour les âges de 45 à 60')
plt.xlabel('Âge')
plt.ylabel('Médiane de ConvertedComp')
plt.show()

import matplotlib.pyplot as plt

# Requête pour obtenir l'âge, la médiane des revenus convertis pour le groupe d'âge 25-30
age_salary_query = """
SELECT Age, AVG(ConvertedComp) as MedianConvertedComp
FROM master
WHERE Age BETWEEN 25 AND 30
GROUP BY Age
"""

# Exécutez la requête et lisez les résultats dans un DataFrame
age_salary_data = pd.read_sql_query(age_salary_query, conn)

# Créer le graphique linéaire
plt.figure(figsize=(10, 6))
plt.plot(age_salary_data['Age'], age_salary_data['MedianConvertedComp'], marker='o', linestyle='-')
plt.title('Line Chart of Age and Median ConvertedComp for Age Group 25-30')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.show()

```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# Horizontal bar chart using column MainBranch
bar_chart_query = """
SELECT MainBranch, COUNT(*) as Count
FROM master
WHERE MainBranch IS NOT NULL
GROUP BY MainBranch
"""

bar_chart_data = pd.read_sql_query(bar_chart_query, conn)

bar_chart_data.plot(kind='barh', x='MainBranch', y='Count')
plt.title('Diagramme à barres horizontales de MainBranch')
plt.xlabel('Comptage')
plt.ylabel('MainBranch')
plt.show()

```
